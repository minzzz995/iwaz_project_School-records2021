```python
import pandas as pd
import numpy as np
import re
import openpyxl
```

```python
wb = openpyxl.load_workbook(r'./07.xlsx')
ws_names = wb.sheetnames
```

```python
dic = list()
dic_check = {}
swap = {1:"A",2:"B",3:"C",4:"D",5:"E",6:'F',7:'G',8:'H',9:'I',10:'J'}
for i in range(1,11):
    data = pd.read_excel('3차유의어정리_기재금지어목록.xlsx',sheet_name = i)
    dic = (list(data["유의어"])) 
    for j in range(len(dic)):
        dic_check[str(dic[j])] = swap.get(i)
```

```python
dic = list()
for i in range(1,11):
    data = pd.read_excel('3차유의어정리_기재금지어목록.xlsx',sheet_name = i)
    dic.append(list(data["유의어"])) 
dic = sum(dic, [])
dic.remove('중')
dic.remove('줌')
dic.remove('배우')
dic.remove('형')
```

```python
sheet_num = 3
data = pd.read_excel('./07.xlsx',sheet_name = sheet_num)
check = list(data.iloc[:,1])
original_sentence = []
words = []
word = []
sheet_name = []
sheet_type = []
val = []
original_num = []
```

```python
for i in range(len(check)): 
    for j in range(len(dic)): 
        if dic[j] in check[i]: 
            check[i] = re.sub(dic[j],"[[[ %s ]]]" % dic[j],check[i])
            word.append(dic[j])
            val.append(dic_check.get(dic[j]))
    for j in range(len(dic)):
        if dic[j] in check[i]:
            original_sentence.append(check[i])
            sheet_name.append(ws_names[sheet_num])
            a = "07_" + ws_names[sheet_num] + "_" + str(i+1)
            original_num.append(a)
            break
    word = re.sub('[\[\]\'\"]',"",str(word))
    val = re.sub('[\[\]\'\"]',"",str(val))
    if word != [] and word != '':
        words.append(str(word))
        sheet_type.append(str(val))
    word = []
    val = []
```

```python
df = pd.DataFrame({"순번":np.nan,"원문 출처":original_num,"원문 문장":original_sentence,
                   "유의어 문장":np.nan,"유의어":words,"항목":sheet_name,"사유구분":sheet_type})
```
